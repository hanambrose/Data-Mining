```
X = np.loadtxt ('X.csv', delimiter=",")
y = np.loadtxt ('y.csv', delimiter=",")
svalues = [0.2,0.4,0.8,1.0,1.2,1.4,1.6]
alpha = LA.inv(X.T @ X) @ X.T @ y.T
print ('Least Squares error %.2f'% LA.norm(y -X @ alpha,2))

print ('Ridge Regression errors:')
svalues = [0.2,0.4,0.8,1.0,1.2,1.4,1.6]
for s in svalues:
  alphas = LA.inv(X.T @ X + s*np.identity(50)) @ X.T @ y.T
  print('s = ', s, ': %.2f'% LA.norm(y -X @ alphas,2))
```

```
#Set 1
X1 = X[:66,:]
y1 = y[:66]
#Set 2
X2 = X[33:,:]
y2 = y[33:]
#Set 3
X3 = np.vstack((X[:33,:], X[66:,:]))
y3 = np.concatenate((y[:33], y[66:]), axis =0)

X3.shape
```

```
print('Least Squares error:')
alpha_1 = LA.inv(X1.T @ X1) @ X1.T @ y1.T # X[:66,:], y[:66] is training set
resid_1 = LA.norm(y[66:] - X[66:,:] @ alpha_1,2) # X[66:,:], y[66:] is test set
print ('Subset 1 %.2f'% resid_1)

alpha_2 = LA.inv(X2.T @ X2) @ X2.T @ y2.T
resid_2 = LA.norm(y[:33] - X[:33,:] @ alpha_2,2)
print ('Subset 2 %.2f'% resid_2)

alpha_3 = LA.inv(X3.T @ X3) @ X3.T @ y3.T
resid_3 = LA.norm(y[33:66] - X[33:66,:] @ alpha_3,2)
print ('Subset 3 %.2f'% resid_3)

print('average LS error %.2f' %  ((resid_1+resid_2+resid_3)/3))
```

```
print("Subset 1")
err_1 = []
for s in svalues:
  alphas = LA.inv(X1.T @ X1 + s*np.identity(50)) @ X1.T @ y1.T
  resid = LA.norm(y[66:] - X[66:,:] @ alphas,2)
  err_1.append(resid)
  print('s = ', s, ':%.2f' % resid)

print("Subset 2")
err_2 = []
for s in svalues:
  alphas = LA.inv(X2.T @ X2 + s*np.identity(50)) @ X2.T @ y2.T
  resid = LA.norm(y[:33] - X[:33,:] @ alphas,2)
  err_2.append(resid)
  print('s = ', s, ':%.2f' % resid)

print("Subset 3")
err_3 = []
for s in svalues:
  alphas = LA.inv(X3.T @ X3 + s*np.identity(50)) @ X3.T @ y3.T
  resid = LA.norm(y[33:66] - X[33:66,:] @ alphas,2)
  err_3.append(resid)
  print('s = ', s, ':%.2f' %resid)

print('Average error')
avg_error_rr = np.average(np.stack((err_1, err_2, err_3)), axis = 0)
print(avg_error_rr)
```

```
sqr_error_ind = [e**2/100 for e in avg_error_rr]
np.round(sqr_error_ind,2)
#sqr_error_ind
```
