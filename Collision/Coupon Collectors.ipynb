```
import random
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
import time
```

```
def coupon(size):
    random_numbers_set = set()
    trials = 0
    while True:
        nextRan = random.randint(1,size)
        trials += 1
        #print('%d nextRan'%nextRan)
        if nextRan not in random_numbers_set:
            random_numbers_set.add(nextRan)
            #print(random_numbers_set)
            if len(random_numbers_set) == size:# continue until we get all types of coupons
                return trials
    return trials
```

```
n = 300
print('%d trials to get all types of coupons'%coupon(n))
### expected
print('when the expected amount of trials is %d'%((math.log(n) + 0.5772)*n))
```

```
n_set= (5000,10000,15000,20000) #number of domain
t=[]
for n in n_set:
  start_time = time.time()
  p = []
  k = []
  m = 2000
  prob = 1/m
  while m > 0:
      nextRand = coupon(n)
      k.append(nextRand)
      m -= 1
  k = sorted(k)
  for z in range(len(k)):
      p.append(k.index(k[z])*prob)
  t.append(time.time() - start_time)    
  #print('%d average trials to get all types of coupons'% np.mean(k))
  #print('when the expected amount of trials is %d'% (n*(0.577+math.log(n)))) 
print (t)
sns.set_style("darkgrid")
#mpl.rcParams['figure.figsize'] = (6,4)
mpl.rcParams['figure.dpi'] = 100
plt.title("runtime with m = 2000")
plt.xlabel("n")
plt.xticks(rotation=45)
plt.ylabel("time in seconds")
plt.scatter (np.array(n_set), np.array(t)) 
plt.show()
```

```
sns.set_style("darkgrid")
mpl.rcParams['figure.figsize'] = (6,4)
mpl.rcParams['figure.dpi'] = 200
plt.title("cummulative density plot given k with n = %d" %n)
plt.xlabel("k")
plt.ylabel("probability of a collision")
plt.plot(np.array(k), np.array(p))
plt.show()
```
