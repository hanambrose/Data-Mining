```
def two_gram_char (data_set):
  with open(data_set,'r') as data_set:
    text = data_set.read()
    k_gram =set()
    for i in range(len(text)-1):
      if (text[i] + text[i+1]) not in k_gram:
        k_gram.add(text[i] + text[i+1])
  return k_gram
```

```
def three_gram_char (data_set):
  with open(data_set,'r') as data_set:
    text = data_set.read()
    k_gram =set()
    for i in range(len(text)-2):
      if (text[i] + text [i+1] + text[i+2]) not in k_gram:
        k_gram.add(text[i] + text [i+1] + text[i+2])
  return k_gram
```

```
def two_gram_word (data_set):
  with open(data_set, 'r') as data_set:
    word = str.split(data_set.read())
    k_gram = set()
    for i in range(len(word)-1):
      if (word[i]+ ' ' + word[i+1]) not in k_gram:
          k_gram.add(word[i]+ ' ' + word[i+1])      
  return k_gram
```

```
doc_set = ['D1.txt','D2.txt','D3.txt','D4.txt']
for data_set in doc_set:
  print(data_set + ' : 2_grams based on character: %d' % len(two_gram_char(data_set)))
```

```
doc_set = ['D1.txt','D2.txt','D3.txt','D4.txt']
for data_set in doc_set:
  print(data_set + ' : 3_grams based on character: %d' % len(three_gram_char(data_set)))
```

```
doc_set = ['D1.txt','D2.txt','D3.txt','D4.txt']
for data_set in doc_set:
  print(data_set + ' : 2_grams based on word: %d' %len(two_gram_word(data_set)))
```

```
def jac_sim(d1,d2,d3,d4,type):
  print(type +' between D1 and D2 is %.5f' % (len(d1.intersection(d2))/len(d1.union(d2))))
  print(type +' between D1 and D3 is %.5f' % (len(d1.intersection(d3))/len(d1.union(d3))))
  print(type +' between D1 and D4 is %.5f' % (len(d1.intersection(d4))/len(d1.union(d4))))
  print(type +' between D2 and D3 is %.5f' % (len(d2.intersection(d3))/len(d2.union(d3))))
  print(type +' between D2 and D4 is %.5f' % (len(d2.intersection(d4))/len(d2.union(d4))))
  print(type +' between D3 and D4 is %.5f' % (len(d3.intersection(d4))/len(d3.union(d4))))
```

```
jac_sim(two_gram_char('D1.txt'), two_gram_char('D2.txt'), two_gram_char('D3.txt'), two_gram_char('D4.txt'),'2-grams char')
print('')
jac_sim(three_gram_char('D1.txt'), three_gram_char('D2.txt'), three_gram_char('D3.txt'), three_gram_char('D4.txt'),'3-gram char')
print('')
jac_sim(two_gram_word('D1.txt'), two_gram_word('D2.txt'), two_gram_word('D3.txt'), two_gram_word('D4.txt'),'2-gram word')
```

```
D1_gram = three_gram_char('D1.txt')
D2_gram = three_gram_char('D2.txt')
D_total = list(D1_gram.union(D2_gram))

for k in [20,60,150,300,600]:    
    coll = 0
    for i in range (k):
        v = [math.inf, math.inf]
        for j in range (len(D_total)):
            h = hash(str(i)+D_total[j]+str(i)) % 10000
            if D_total[j] in D1_gram: 
                if (h < v[0]):
                    v[0] = h
            if D_total[j] in D2_gram: 
                if (h < v[1]):
                    v[1] = h
        if v[0] == v[1]:
            coll = coll+1
    print("By using MinHash, t = %d"%k, ": approximate JS between D1 and D2 is ", coll/k)
```
